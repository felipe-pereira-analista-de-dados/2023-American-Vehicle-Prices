# Sobre o conjunto de dados
#### Estes dados foram extraídos de https://www.ccarprice.com/us/ em 12 de setembro de 2023. Os dados mostraram todos modelos anunciados em 2023, não apenas em anos modelo de 2023, então qualquer coisa não marcada como 2023 foi filtrada. Os preços estão em dólares americanos (USD). Para alguns veículos, estão presentes vários preços, pelo que pode ser calculada uma gama.






```python
# importação das bibliotecas
import pandas as pd

# importação do arquivo de dados
tabela = pd.read_csv("..\Data\prices_clean.csv")
display(tabela)
```

```python
# Quantidade de marcas diferentes na tabela
soma_marca = tabela['make'].nunique()
print(f'Quantidade de marcas diferentes: {soma_marca}')
```

```python
# ordem decrescente dos valores de carros
revenue_brand = tabela[['make', 'price']].groupby('make').sum().sort_values(by='price', ascending=False)
revenue_brand.head(10)
```
